## *Question 2*

2-Définition d'une fonction nommée "compute_error" en python qui prend en argument a, b, X, Y et qui 
calcule J (a, b) la fonction du coût (ensemble des erreurs):

La fonction 'compute_error(b,a,points)' parcourt chaque point de l'ensemble de données. Pour chaque point, elle détermine la valeur prédite de y sur la ligne définie par l'équation y = ax + b . En comparant cette valeur prédite avec la valeur réelle de y, elle calcule l'erreur quadratique. Ces erreurs quadratiques sont ensuite accumulées et la fonction retourne leur moyenne. 

```python
#
import numpy as np
import matplotlib.pyplot as plt

def compute_error(b,a,points):
    totalerror=0
    for i in range(0,len(points)):
        x=points[i,0]
        y=points[i,1]
        totalerror+=(y-(a*x+b))**2
    return totalerror/(2*float(len(points)))
```

## *Question 3*

3- Définir une fonction nommée "calcul_gradient" en python, qui calcule les nouveaux a et b mises à 
jour en utilisant les dérivées partielles. 

```python
#
def calcul_gradient(b_current,a_current,points,learnrate):
    b_gradient=0
    a_gradient=0
    N=float(len(points))

    for i in range(0,len(points)):
        x=points[i,0]
        y=points[i,1]

        b_gradient+=(1/N)*((a_current*x+b_current)-y)
        a_gradient+=(1/N)*x*((a_current*x+b_current)-y)
    #renew
    new_b=b_current-(learnrate*b_gradient)
    new_a=a_current-(learnrate*a_gradient)
    return [new_b,new_a]
```

## *Question 4*

4-Définir une fonction nommée "GD_runner" en python, qui permet de calculer les nouveaux a et b 
pour n itérations. Pour chaque itération, veuillez afficher les éléments suivants : 

A- La valeur de la fonction du coût 

B- La valeur du coefficient a 

C- La valeur du coefficient b 

```python
def GD_runner(points,starting_b,starting_a,learning_rate,num_iteration):
    b=starting_b
    a=starting_a
    cost_history = np.zeros(num_iteration)
    for i in range(num_iteration):
        b,a=calcul_gradient(b,a,np.array(points),learning_rate)
        print("Iteration Numero :",i)
        print("Fonction du cout = ", compute_error(b,a,np.array(points)))
        cost_history[i] = compute_error(b,a,np.array(points))
        print("a=", a)
        print("b=", b)
    plt.plot(range(num_iteration), cost_history)
    plt.show()
    return [b,a]
```

5- Visualiser l'évolution de la fonction du coût J (a, b) au des n itérations.

        Hyperparameter : learning_rate = 0.0001 et nombre_itération = 100

```python
# Hyperparameter : learning_rate = 0.0001 et nombre_itération = 100 :
def run():

    # Initialize one hundred points
    x=np.linspace(1,10,100)
    y=np.linspace(1,10,100)

    array=[]

    for i in np.arange(100):
        for j in np.arange(100):
            p=[x[i],y[i]]
            array.append(p)


    learning_rate=0.0001
    init_b=0
    init_a=0
    num_iteration=100

    [b,a]=GD_runner(array,init_b,init_a,learning_rate,num_iteration)

    plt.figure()

    plt.scatter(x, y)
    y2=b+a*x

    plt.scatter(x,y2)
    print("last_b=", b)
    print("last_a=", a)

if __name__=="__main__":
    run()
```
